```python

import random
print("it is random number guessing game")
print("you need to select number of trails and then enter the guess value")
print("you need to enter numbers only")
rand_num=random.randint(1,10)
trails=int(input("no. Of trails you want to perform"))
for i in range(trails):
    user_num=int(input("enter the number to guess"))
    if rand_num==user_num:
        print("you won")
        break
    elif trails-1-i==0:
        print("you lost play after some time")
    else:
         print("you lost")
         print("the chances left are",trails-1-i)
         print("------------*****-----------")
```
